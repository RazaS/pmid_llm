```python
from Bio import Entrez
import pandas as pd

# Function to search PubMed
def search(query, retmax=10):
    Entrez.email = 'your_email@example.com'  # Replace with your email
    handle = Entrez.esearch(
        db='pubmed',
        sort='relevance',
        retmax=retmax,
        retmode='xml',
        term=query
    )
    results = Entrez.read(handle)
    handle.close()
    return results['IdList']

# Function to fetch article details
def fetch_details(id_list):
    ids = ','.join(id_list)
    Entrez.email = 'your_email@example.com'
    handle = Entrez.efetch(
        db='pubmed',
        retmode='xml',
        id=ids
    )
    results = Entrez.read(handle)
    handle.close()
    return results

# Main script to get abstracts
def fetch_abstracts(query, max_results=10):
    pmids = search(query, retmax=max_results)
    abstracts = []
    if pmids:
        papers = fetch_details(pmids)
        for paper in papers['PubmedArticle']:
            try:
                abstract = paper['MedlineCitation']['Article']['Abstract']['AbstractText'][0]
            except KeyError:
                abstract = "No Abstract Available"
            abstracts.append(abstract)
    return abstracts

if __name__ == "__main__":
    query = "transfusion AND free full text[Filter]"
    abstracts = fetch_abstracts(query, max_results=10)
    
    # Display the results
    for i, abstract in enumerate(abstracts, 1):
        print(f"Abstract {i}:\n{abstract}\n")
```
